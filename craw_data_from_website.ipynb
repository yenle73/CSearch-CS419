```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://openaccess.thecvf.com/WACV2023'
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
page
```

```python
soup
```

```python
titles = []
papers = soup.find_all('dt', class_='ptitle')
titles += [paper.find('a').text for paper in papers]
```

```python
titles[0]
```

```python
titles.index('3D Change Localization and Captioning From Dynamic Scans of Indoor Scenes')
```

```python
import xml.etree.ElementTree as ET
```

```python
root = ET.Element("root")
for title in range(len(titles)):
  doc = ET.SubElement(root, "doc")
  doc.text = str(titles[title])

for i, doc in enumerate(root.findall("doc")):
  doc.set("id", str(i))
```

```python
tree = ET.ElementTree(root)
tree.write("papers.xml")
```

```python
ET.dump(root)
```

```python
link_text = []
```

```python
for a in soup.find_all('a', href=True, string=True):
  if re.search('paper.pdf', a['href']):
    link_text.append(a['href'])
```

```python
for link in link_text:
  link = 'https://openaccess.thecvf.com/' + link
```

```python
root = ET.Element("root")
for title in range(len(link_text)):
  link = ET.SubElement(root, "link")
  link.text = str(link_text[title])

for i, link in enumerate(root.findall("link")):
  link.set("id", str(i))
```

```python
tree = ET.ElementTree(root)
tree.write("links_fix.xml")
```
